```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
logistic = np.random.logistic(1,3,100) #logistic(location, scale, repetitions)
plt.hist(logistic, color = 'red')
plt.axvline(logistic.mean())
plt.axvline(logistic.mean() + logistic.std(), linestyle = 'dashed')
plt.axvline(logistic.mean() - logistic.std(), linestyle = 'dashed')
plt.show()
print("Mean: " + str(logistic.mean()))
print("Standard dev: " + str(logistic.std()))
```

```python
lognormal = np.random.lognormal(1,3,100) #lognormal(mean, sigma, repetitions)
plt.hist(lognormal, color = 'green')
plt.axvline(lognormal.mean())
plt.axvline(lognormal.mean() + lognormal.std(), linestyle = 'dashed')
plt.axvline(lognormal.mean() - lognormal.std(), linestyle = 'dashed')
plt.show()
print("Mean: " + str(lognormal.mean()))
print("Standard dev: " + str(lognormal.std()))
```

```python
uniform = np.random.uniform(1,3,100)
plt.hist(uniform, color='yellow')
plt.axvline(uniform.mean())
plt.axvline(uniform.mean()+uniform.std(), linestyle='dashed')
plt.axvline(uniform.mean()-uniform.std(), linestyle='dashed')
plt.show()
print("Mean: " + str(uniform.mean()))
print("Standard dev: " + str(uniform.std()))
```

```python
wald = np.random.wald(1,3,100)
plt.hist(wald, color='orange')
plt.axvline(wald.mean())
plt.axvline(wald.mean()+wald.std(), linestyle='dashed')
plt.axvline(wald.mean()-wald.std(), linestyle='dashed')
plt.show()
print("Mean: " + str(wald.mean()))
print("Standard dev: " + str(wald.std()))
```

```python
triangular = np.random.triangular(1,3,20,100)
plt.hist(triangular)
plt.axvline(triangular.mean(), color='red')
plt.axvline(triangular.mean()+triangular.std(), color='red', linestyle='dashed')
plt.axvline(triangular.mean()-triangular.std(), color='red', linestyle='dashed')
plt.show()
print("Mean: " + str(triangular.mean()))
print("Standard dev: " + str(triangular.std()))
```

```python
exponential = np.random.exponential(3,100)
plt.hist(exponential, color='purple')
plt.axvline(exponential.mean())
plt.axvline(exponential.mean()+exponential.std(),linestyle='dashed')
plt.axvline(exponential.mean()-exponential.std(),linestyle='dashed')
plt.show()
print("Mean: " + str(exponential.mean()))
print("Standard dev: " + str(exponential.std()))
```

Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with a mean of 10 and standard deviation of 1.
Add them together to create a third variable.
Graph the third variable using a histogram.
Compute the mean and standard deviation and plot them as vertical lines on the histogram.
Evaluate the descriptive statistics against the data.

```python
norm1 = np.random.normal(5, .5,100)
norm2 = np.random.normal(10, 1,100)
norm3 = norm1 + norm2
plt.hist(norm3, color='pink')
plt.axvline(norm3.mean())
plt.axvline(norm3.mean()+norm3.std(), linestyle='dashed')
plt.axvline(norm3.mean()-norm3.std(), linestyle='dashed')
plt.plot()
```
