```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mean = 0
std = 1
n = 1000
```

```python
df = pd.DataFrame({'rand': np.random.normal(mean, std, n)})
```

```python
df.hist()
plt.show()
```

```python
bern_dist = np.random.binomial(1,.5, 100) #what is the result of an event done n times, with prob p of success, tested "size" times
#In Bernouli distribution, n=1
print(bern_dist)
```

```python
plt.hist(bern_dist)
plt.show()
```

```python
gamma = np.random.gamma(10,1,1000)
```

```python
plt.hist(gamma)
plt.show()
```

```python
poisson = np.random.poisson(40,1000)
```

```python
plt.hist(poisson)
plt.show()
```

```python
ecommerce = pd.DataFrame()
ecommerce['spending'] = np.random.normal(75,25,1000)
```

```python
ecommerce.hist()
plt.show()
```

```python
ecommerce['visit_count'] = np.random.randint(0,5,1000)
```

```python
plt.hist(ecommerce[ecommerce['visit_count']>2]['spending'])
plt.show()
```
